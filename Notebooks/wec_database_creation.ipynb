```python
file_paths = [
    '../Datasets/2014/Silverstone_2014.csv',
    '../Datasets/2014/Spa_2014.csv',
    '../Datasets/2014/Bahrain_2014.csv',
    '../Datasets/2014/Cota_2014.csv',
    '../Datasets/2014/Fuji_2014.csv',
    '../Datasets/2014/Interlagos_2014.csv',
    '../Datasets/2014/LeMans_2014.csv',
    '../Datasets/2014/Shanghai_2014.csv',
    '../Datasets/2015/Silverstone_2015.csv',
    '../Datasets/2015/Spa_2015.csv',
    '../Datasets/2015/Bahrain_2015.csv',
    '../Datasets/2015/Cota_2015.csv',
    '../Datasets/2015/Fuji_2015.csv',
    '../Datasets/2015/Nurburgring_2015.csv',
    '../Datasets/2015/LeMans_2015.csv',
    '../Datasets/2015/Shanghai_2015.csv',
    '../Datasets/2016/Silverstone_2016.csv',
    '../Datasets/2016/Spa_2016.csv',
    '../Datasets/2016/Bahrain_2016.csv',
    '../Datasets/2016/Cota_2016.csv',
    '../Datasets/2016/Fuji_2016.csv',
    '../Datasets/2016/Nurburgring_2016.csv',
    '../Datasets/2016/LeMans_2016.csv',
    '../Datasets/2016/Shanghai_2016.csv',
    '../Datasets/2016/Mexico_2016.csv',
    '../Datasets/2017/Silverstone_2017.csv',
    '../Datasets/2017/Spa_2017.csv',
    '../Datasets/2017/Bahrain_2017.csv',
    '../Datasets/2017/Cota_2017.csv',
    '../Datasets/2017/Fuji_2017.csv',
    '../Datasets/2017/Nurburgring_2017.csv',
    '../Datasets/2017/LeMans_2017.csv',
    '../Datasets/2017/Shanghai_2017.csv',
    '../Datasets/2017/Mexico_2017.csv',
    '../Datasets/2018-2019/Silverstone_2018-2019.csv',
    '../Datasets/2018-2019/Spa_2018-2019.csv',
    '../Datasets/2018-2019/Fuji_2018-2019.csv',
    '../Datasets/2018-2019/LeMans_2018-2019.csv',
    '../Datasets/2018-2019/LeMans2_2018-2019.csv',
    '../Datasets/2018-2019/Shanghai_2018-2019.csv',
    '../Datasets/2018-2019/Sebring_2018-2019.csv',
    '../Datasets/2018-2019/Spa2_2018-2019.csv',
    '../Datasets/2019-2020/Silverstone_2019-2020.csv',
    '../Datasets/2019-2020/Fuji_2019-2020.csv',
    '../Datasets/2019-2020/Shanghai_2019-2020.csv',
    '../Datasets/2019-2020/Bahrain_2019-2020.csv',
    '../Datasets/2019-2020/Cota_2019-2020.csv',
    '../Datasets/2019-2020/LeMans_2019-2020.csv',
    '../Datasets/2019-2020/Spa_2019-2020.csv',
    '../Datasets/2019-2020/Bahrain_2019-2020.csv',
    '../Datasets/2021/Algarve_2021.csv',
    '../Datasets/2021/LeMans_2021.csv',
    '../Datasets/2021/Spa_2021.csv',
    '../Datasets/2021/Monza_2021.csv',
    '../Datasets/2021/Bahrain_2021.csv',
    '../Datasets/2021/Bahrain2_2021.csv',
    '../Datasets/2022/Sebring_2022.csv',
    '../Datasets/2022/LeMans_2022.csv',
    '../Datasets/2022/Spa_2022.csv',
    '../Datasets/2022/Monza_2022.csv',
    '../Datasets/2022/Fuji_2022.csv',
    '../Datasets/2022/Bahrain_2022.csv',
    '../Datasets/2023/Bahrain_2023.csv',
    '../Datasets/2023/Algarve_2023.csv',
    '../Datasets/2023/LeMans_2023.csv',
    '../Datasets/2023/Spa_2023.csv',
    '../Datasets/2023/Monza_2023.csv',
    '../Datasets/2023/Fuji_2023.csv',
    '../Datasets/2023/Sebring_2023.csv',
    '../Datasets/2011/LeMans_2011.csv',
    '../Datasets/2012/Silverstone_2012.csv',
    '../Datasets/2012/Spa_2012.csv',
    '../Datasets/2012/Bahrain_2012.csv',
    '../Datasets/2012/Shanghai_2012.csv',
    '../Datasets/2012/Fuji_2012.csv',
    '../Datasets/2012/Interlagos_2012.csv',
    '../Datasets/2012/LeMans_2012.csv',
    '../Datasets/2012/Sebring_2012.csv',
    '../Datasets/2013/Silverstone_2013.csv',
    '../Datasets/2013/Spa_2013.csv',
    '../Datasets/2013/Bahrain_2013.csv',
    '../Datasets/2013/Shanghai_2013.csv',
    '../Datasets/2013/Fuji_2013.csv',
    '../Datasets/2013/Interlagos_2013.csv',
    '../Datasets/2013/LeMans_2013.csv',
    '../Datasets/2013/Cota_2013.csv',
       
]

dfs = []

for file_path in file_paths:
    df = pd.read_csv(file_path, delimiter=';')
    race = file_path.split('/')[-1].split('_')[0]
    season = str(file_path.split('_')[-1].split('.')[0])
    df['race'] = race
    df['season'] = season
    dfs.append(df)


wec_data = pd.concat(dfs, ignore_index=True)
```

```python
# Rename all columns to lowercase
wec_data.columns = [col.lower() for col in wec_data.columns]

# Rename the 'position' column to 'overall_position'
wec_data.rename(columns={'position': 'overall_position'}, inplace=True)

# Rename the 'fl_lapnum' column to 'fl_lap_number'
wec_data.rename(columns={'fl_lapnum': 'fl_lap_number'}, inplace=True)

# Rename the 'fl_kph' column to 'fl_kph_average'
wec_data.rename(columns={'fl_kph': 'fl_kph_average'}, inplace=True)

# Rename the 'gap_previous' column to 'gap_car_ahead'
wec_data.rename(columns={'gap_previous': 'gap_car_ahead'}, inplace=True)
```

```python
# Concatenate 'number' and 'team' columns as a new column 'car'
wec_data['car'] = wec_data['team'] + ' #' + wec_data['number'].astype(str)
```

```python

# Map 'D' to 'Dunlop', 'M' to 'Michelin', and 'G' to 'Goodyear' in the 'Tyres' column
wec_data['tyres'] = wec_data['tyres'].map({'D': 'Dunlop', 'M': 'Michelin', 'G': 'Goodyear', 'P': 'Pirelli', 'H': 'Hankook'})
```

```python
# Drop the 'Unnamed: 20' column
wec_data = wec_data.drop(['unnamed: 20', 'driver_4', 'division', 'number'], axis=1)
```

```python
# Create a new column 'class_position' based on 'overall_position' within each 'class'
wec_data['class_position'] = wec_data.groupby(['season', 'race', 'class'])['overall_position'].rank().astype(int)
```

```python
# Replace single quotes with colons in the 'total_time' and 'fl_time' columns
wec_data['total_time'] = wec_data['total_time'].str.replace("'", ":")
wec_data['fl_time'] = wec_data['fl_time'].str.replace("'", ":")
```

```python
# Update the condition for Bahrain 2019 to have event_duration as 8
conditions = [
    (wec_data['race'].isin(['LeMans', 'LeMans2'])),
    (wec_data['race'] == 'Sebring') & (wec_data['season'].isin(['2018-2019', '2022', '2023'])),
    (wec_data['race'].isin(['Silverstone', 'Shanghai']) & (wec_data['season'] == '2019-2020')),
    (wec_data['race'] == 'Bahrain') & (wec_data['season'].isin(['2019-2020', '2021', '2022'])),
    (wec_data['race'] == 'Algarve') & (wec_data['season'].isin(['2021'])),
    (wec_data['race'] == 'Sebring') & (wec_data['season'].isin(['2012']))
]

choices = [24, 8, 4, 8, 8, 12]

wec_data['event_duration'] = np.select(conditions, choices, default=6)
```

```python
# Rename races that occured twice in a season
wec_data['race'] = wec_data['race'].replace('Bahrain2', 'Bahrain 2')
wec_data['race'] = wec_data['race'].replace('Spa2', 'Spa 2')
wec_data['race'] = wec_data['race'].replace('LeMans2', 'LeMans 2')
```

```python
# Reorder columns as needed
wec_data = wec_data[['car', 'overall_position', 'class_position', 'team', 'vehicle', 'class', 'group', 'race', 'event_duration', 'season', 'laps', 'total_time', 'gap_first', 'gap_car_ahead', 'tyres', 'status', 'fl_lap_number', 'fl_time', 'fl_kph_average', 'driver_1', 'driver_2', 'driver_3']]  
```

```python
# Write the merged dataframe to a new CSV file
wec_data.to_csv('../Datasets/wec_data.csv', index=False)
```
