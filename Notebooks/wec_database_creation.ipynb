```python
# Rename all columns to lowercase
wec_data.columns = [col.lower() for col in wec_data.columns]

# Rename the 'position' column to 'overall_position'
wec_data.rename(columns={'position': 'overall_position'}, inplace=True)

# Rename the 'fl_lapnum' column to 'fl_lap_number'
wec_data.rename(columns={'fl_lapnum': 'fl_lap_number'}, inplace=True)

# Rename the 'fl_kph' column to 'fl_kph_average'
wec_data.rename(columns={'fl_kph': 'fl_kph_average'}, inplace=True)

# Rename the 'gap_previous' column to 'gap_car_ahead'
wec_data.rename(columns={'gap_previous': 'gap_car_ahead'}, inplace=True)
```

```python
# Concatenate 'number' and 'team' columns as a new column 'car'
wec_data['car'] = wec_data['team'] + ' #' + wec_data['number'].astype(str)
```

```python

# Map 'D' to 'Dunlop', 'M' to 'Michelin', and 'G' to 'Goodyear' in the 'Tyres' column
wec_data['tyres'] = wec_data['tyres'].map({'D': 'Dunlop', 'M': 'Michelin', 'G': 'Goodyear'})
```

```python
# Drop the 'Unnamed: 20' column
wec_data = wec_data.drop(['unnamed: 20', 'driver_4', 'division', 'number'], axis=1)
```

```python
# Create a new column 'class_position' based on 'overall_position' within each 'class'
wec_data['class_position'] = wec_data.groupby(['season', 'race', 'class'])['overall_position'].rank().astype(int)
```

```python
# Replace single quotes with colons in the 'total_time' and 'fl_time' columns
wec_data['total_time'] = wec_data['total_time'].str.replace("'", ":")
wec_data['fl_time'] = wec_data['fl_time'].str.replace("'", ":")
```

```python
# Update the condition for Bahrain 2019 to have event_duration as 8
conditions = [
    (wec_data['race'].isin(['LeMans', 'LeMans2'])),
    (wec_data['race'] == 'Sebring') & (wec_data['season'].isin(['2018-2019', '2022', '2023'])),
    (wec_data['race'].isin(['Silverstone', 'Shanghai']) & (wec_data['season'] == '2019-2020')),
    (wec_data['race'] == 'Bahrain') & (wec_data['season'].isin(['2019-2020', '2021', '2022'])),
    (wec_data['race'] == 'Algarve') & (wec_data['season'].isin(['2021'])),
    (wec_data['race'] == 'Sebring') & (wec_data['season'].isin(['2012']))
]

choices = [24, 8, 4, 8, 8, 12]

wec_data['event_duration'] = np.select(conditions, choices, default=6)
```

```python
# Rename races that occured twice in a season
wec_data['race'] = wec_data['race'].replace('Bahrain2', 'Bahrain 2')
wec_data['race'] = wec_data['race'].replace('Spa2', 'Spa 2')
wec_data['race'] = wec_data['race'].replace('LeMans2', 'LeMans 2')
```

```python
# Reorder columns as needed
wec_data = wec_data[['car', 'overall_position', 'class_position', 'team', 'vehicle', 'class', 'group', 'race', 'event_duration', 'season', 'laps', 'total_time', 'gap_first', 'gap_car_ahead', 'tyres', 'status', 'fl_lap_number', 'fl_time', 'fl_kph_average', 'driver_1', 'driver_2', 'driver_3']]  
```
